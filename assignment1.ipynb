## 3. iris의 데이터 타입을 np.float64로 바꾸고, 바꾼 후의 iris의 데이터 타입을 프린트하세요.

```python
iris = iris.astype(np.float64)
iris.dtype
```

## 4. 유니버설 함수를 활용해서 iris의 소수부를 모두 버리고, 수정된 iris를 프린트하세요.

```python
iris = np.floor(iris)
iris
```

## 5. iris와 같은 크기 (150,4) 이면서 표준정규분포를 따르는 noise 배열을 만들고 프린트하세요. (넘파이 random 모듈의 난수 generator를 생성하여 활용하고, 그 seed는 1000번으로 하세요.)

```python
rng = np.random.default_rng(seed=1000)
noise = rng.standard_normal((150, 4))
noise
```

## 6. iris와 noise를 더한뒤, 1번 axis방향으로 누적합으로 iris를 수정하고, 프린트하세요. (크기가 (150,4)인 넘파이 배열)

```python
iris = np.cumsum(iris+noise, axis=1)
iris
```

```python
iris = np.where(iris>10, 10, iris) # 10이 넘어가면 0으로 수정, 안 넘어가면 기존 값 유지
iris
```

```python
import pandas as pd
from pandas import Series, DataFrame
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='str')
```

## 8. column을 순서대로 'a','b','c','d','e' 로 갖는 데이터프레임의 형태로 iris를 만들고 프린트하세요.

```python
iris = DataFrame(iris, columns=["a", "b", "c", "d", "e"])
iris
```

## 9. iris의 index에 "samples', columns에 'information'이라는 name 속성을 주고, iris를 프린트하세요.

```python
iris.index.name="samples"
iris.columns.name="information"
iris
```

## 10. reindex를 활용하여, iris의 columns 중에서는 'a','c','e' 와 index는 짝수만 갖도록 수정하고, 프린트하세요.

```python
iris = iris.reindex(columns=["a", "c", "e"], index=np.arange(0,150,2))
iris
```

## 11. iris의 'a' 피처와 'c' 피처를 astype을 활용하여, np.float64로 형변환을 진행하고, loc을 활용하여, 'a'피처의 값이 5를 넘는 sample들만 남도록하세요, 수정된 iris를 프린트하세요.


```python
iris[['a', 'c']] = iris[['a', 'c']].astype(np.float64)
```

```python
iris = iris.loc[iris.a>5, :] # iris = iris.loc[iris['a']>5]
iris
```

## 12. reindex를 활용해서 iris의 index를 0부터 149까지로 수정하고, columns는 a와 c만 남겨놓으세요. 결측치는 그 이전 값으로 채우세요.

```python
iris = iris.reindex(index=np.arange(150), columns=['a', 'c'], method='ffill')
iris
```

## 13. apply 메서드를 활용해서, 각 sample 별로 'a' 피처와 'c' 피처의 평균값과, 둘 중 더 큰 값을 각각 'mean'과 'max'라는 이름의 피처로 갖도록 하는 데이터 프레임을 만드세요.

```python
# iris.apply(lambda x: Series([np.mean(x), np.max(x)], index=['mean', 'max']), axis='columns')
mean_max = DataFrame(index=iris.index)
mean_max['mean'] = iris[['a', 'c']].apply(np.mean, axis=1)
mean_max['max'] = iris[['a', 'c']].apply(np.max, axis=1)

mean_max
```
