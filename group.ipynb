## 데이터 집계
- 배열로부터 스칼라값을 만들어내는 모든 데이터 변환 작업
- 앞선 메소드 중 mean, count, min, sum 등은 집계의 일종
- 각 메소드는 그룹별로 메소드가 적용됨
- aggregate or agg는 임의의 함수를 통해 집계 가능하도록 함
  - 마찬가지로, 그룹별로 함수 적용

### 데이터 집계 관련 다양한 예시
- agg 인수로 메소드 이름 넣을 수 있음
  - ex.agg("mean")
- agg 인수로 메소드 이름 or 함수 배열 넣을 수 있음
  - ex. agg(["mean", "std", "peak_to_peak"])
- 튜플 형태로 작성하여 적용되느 함수에 해당하는 열 이름을 적을 수 있음
  - ex.[("Average", "mean"), ("Variance", np.var)]
- 여러 개의 열에 적용 가능
  - ffuples = [("Average", "mean"), ("Variance", np.var)]
  - grouped[["tip_pct", "total_bill"]].agg(ffuples)
- 열마다 다른 함수 적용 가능
  - grouped.agg({"tip_pct": ["min", "max", "mean", "std"], "size": "sum"})
- as_index_False를 통해, 인덱스를 그룹화된 인덱스로 바꾸지 않을 수 있음

## apply 메서드
- groupby를 통해 그룹화된 객체에 대해 "일반적인" 메서드
- groupby의 agg, mean, max 등 집계 방법과 가장 큰 차이는
  - agg, mean, max 등은 피쳐 별로 독립적인 집계를 수행
  - apply는 입력값이 그룹화된 데이터 프레임으로, 모든 행과 열을 동시에 고려
- apply 인수
  - 함수, 함수의 인수
  - apply(top, n=1, column='total_bill')

+ group_keys=False를 통해 인덱싱 작업 하지 않을 수 있음

## 6. apply 메서드

```python
def top(df, n=5, column="tip_pct"):
  return df.sort_values(column, ascending=False)[:n] # tip_pct를 기준으로 내림차순, 상위 6개 행 반환(슬라이스)
top(tips, n=6) # n 덮어쓰기 가능
```

```python
tips.groupby("smoker").apply(top)
```

```python
tips.groupby(["smoker", "day"]).apply(top, n=1, column="total_bill")
```

### group_keys=False를 통해 인덱싱 작업을 하지 않도록 가능

```python
tips.groupby(["smoker", "day"], group_keys=False).apply(top, n=1, column="total_bill")
```

## 7. groupby를 활용한 예시 1 - 사순위수 분석, 버킷 분석

```python
frame = pd.DataFrame({"data1": np.random.standard_normal(1000),
                      "data2": np.random.standard_normal(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame["data1"], 4) # frame["data1"]을 4개의 구간으로 나눔, quartiles는 범주형 데이터
quartiles.head(10)
```

```python
def get_stats(group):
  return pd.DataFrame(
      {"min": group.min(), "max": group.max(),
       "count": group.count(), "mean": group.mean()}
  )
grouped = frame.groupby(quartiles) # pd.cut(or qcut)의 결과인 category를 groupby의 기준으로 삼기 가능
grouped.apply(get_stats)
```

```python
grouped.agg(["min", "max", "count", "mean"])
```

```python
quartiles_samp = pd.qcut(frame["data1"], 4, labels=False) # 4개의 구간(0, 1, 2, 3)으로 나눔

# data1의 값 오름차순 정렬 -> 데이터를 4등분하여 각 구간에 동일한 개수의 값이 들어가도록 나눔
quartiles_samp.head()
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
```

## 9. groupby를 활용한 예시 3 - 랜덤 표본과 순열

```python
suits = ["H", "S", "C", "D"] # Hearts, Spades, Clubs, Diamonds
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ["A"] + list(range(2, 11)) + ["J", "K", "Q"]
cards = []
for suit in suits:
  cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
deck
```

```python
def draw(deck, n=5):
  return deck.sample(n)
draw(deck)
```

```python
def get_suit(card):
  return card[-1] # last letter is suit

deck.groupby(get_suit).apply(draw, n=2)
```

```python
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

## transform - #10
- transform의 결과의 크기는 입력된 그룹과 항상 동일
- 집계 결과를 브로드캐스팅하여 결과를 반환
- apply와 다르게 집계 메소드 이름 입력 가능

## 10. transform

```python
df = pd.DataFrame({'key': ['a', 'b', 'c'] * 4,
                   'value': np.arange(12.)})
df
```

```python
g = df.groupby('key')['value']
g.mean()
```

```python
def get_mean(group):
  return group.mean()
g.transform(get_mean)
```

```python
g.transform('mean')
```

```python
g.apply('mean')
```

```python
g.apply(get_mean)
```

```python
def times_two(group):
  return group * 2
g.transform(times_two)
```

```python
def get_ranks(group):
  return group.rank(ascending=False)
g.transform(get_ranks)
```

```python
def normalize(x):
  return (x - x.mean()) / x.std()
```

```python
g.transform(normalize)
```

```python
g.apply(normalize) # 같은 결과
```

## 피벗 테이블
### pivot_table - #11
- groupby + agg와 같은 역할을 하는 메소드

- 그룹화를 원하는 행, 열의 피쳐를 결정하여 원하는 집계 방법에 따라 데이터를 요약해주는 데이터 프레임 메소드
  - 디폴트는 평균치 (mean)
  - 행: index
  - 열: columns
  - 표시할 피쳐: values
  - 집계 메소드: aggfunc

- margins: 부분합이나 총합을 담기 위한 행열을 추가할지 여부
- margins_name: margins를 사용할 때 행/열의 이름

```python
tips.head()
```

```python
tips.pivot_table(index=["day", "smoker"],
                 values=["size", "tip", "tip_pct", "total_bill"]) # 기본은 평균치 집계
```

```python
tips.pivot_table(index=["time", "day"], columns=["smoker", "size"],
                 values=["tip_pct"])
```

```python
tips.pivot_table(index=["time", "day"], columns="smoker",
                values=["tip_pct", "size"])
```

```python
tips.pivot_table(index=["time", "day"], columns="smoker",
                 values=["tip_pct", "size"], margins=True)
```

```python
tips.pivot_table(index=["time", "smoker"], columns="day",
                 values="tip_pct", aggfunc=len, margins=True)
```

```python
tips.pivot_table(index=["time", "smoker"], columns="day",
                 values="tip_pct", aggfunc="count", margins=True)
```

```python
tips.pivot_table(index=["time", "size", "smoker"], columns="day",
                 values="tip_pct", fill_value=0)
```
