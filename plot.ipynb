### 그래프 종류 (matplotlib, seaborn)
- 선 그래프
  - plot, lineplot
    - 가장 기본적인 그래프로 숫자 데이터 x, y 필요
    - x가 없을 시 정수 인덱스
- 막대 그래프
  - bar, barplot
    - x, y가 필요하고, y는 숫자여야 함
    - y 값을 막대로 나타냄
- 가로 막대 그래프
  - barh, barplot
    - x, y 필요, x는 숫자
    - 숫자를 막대로 나타냄
    - seaborn 에서는 x와 y에 따라 바꿔줌.
- 개수 그래프
  - bar, countplot
    - 중복된 요소는 하나로 카운트하여 막대그래프 형태로 나타냄
    - matplotlib에서는 x별 중복된 y를 하나로 카운트하여 도수를 세는 과정을 처리
    - seaborn에서는 countplot 적용으로 간단함
- 박스 플롯
  - boxplot, boxplot
    - y는 수치
    - 항목별로 최소값, 최대값, 사분위수, 이상치를 나타냄
    - 박스는 사분위수 범위(전체 항목의 50퍼센트)를 나타냄
- 산점도
  - scatter, scatterplot
    - 숫자 x, y 필요
    - 각 쌍을 점으로써 데이터를 나타냄
- 히스토그램
  - hist, histplot
    - 숫자 x 필요
    - 구간별로 도수를 나타냄

### plot(및 다양한 plottype)의 다양한 옵션
- plot 시에 각 그래프별 옵션도 있지만, 공유되는 옵션들도 존재
  - color, label, linestyle, marker, alpha 등
  - 그래프 색깔
  - 그래프 이름(범례 또는 legend)
  - 선 스타일 (dashed, solid 등)
  - 마커 (o, * 등)
  - alpha (투명도, 높을 수록 진함)

- ax.legend(): 범례를 나타냄
  - ax에서 plot할 때, label 옵션이 범례에서의 그래프 이름

```python
# 색상, 마커, 선 스타일
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(np.random.standard_normal(30).cumsum(), color="red", linestyle="solid",
        marker="*", alpha=1)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum() # 데이터 변수에 담음
ax.plot(data, color="black", linestyle="dashed", label="default")
ax.plot(data, color="black", linestyle="solid",
        drawstyle="steps-post", label="steps-post") # 계단식..
ax.legend() # 범례를 나타냄
```

### p.19 문제
아래 두 개의 그래프를 각각 plot 한 두 개의 axes 객체를 (1,2) subplot을 갖는 figure에서 plot하시오.
- rand 을 통해 만든 1000개의 데이터를 cumsum한 그래프 색깔을 검정색, dashed이고 label은 'black'
- 0~1000까지 변하는 정수 그래프, 빨간색, label은 'red'
- subplots의 옵션을 활용해서 두 subplot의 x축 눈금과 y축 눈금을 같도록
- 레전드를 나타내시오


```python
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True)
ax[0].plot(np.random.randn(1000).cumsum(), color="black", linestyle="dashed", label="black")
ax[1].plot(np.arange(1000), color="red", label="red")

ax[0].legend()
ax[1].legend()
```

```python
# generator 객체를 사용하며 seed를 설정해서 항상 동일한 난수로 생성됨.
rng = np.random.default_rng(seed=12345)

fig, ax = plt.subplots(1, 2, sharex=True, sharey=True)
ax[0].plot(rng.standard_normal(1000).cumsum(), color="black", label="black", linestyle="dashed")
ax[1].plot(np.arange(1000), color="red", label="red")

ax[0].legend()
ax[1].legend()
```

```python
tips = pd.read_csv("examples/tips.csv")
tips.head()
```

```python
tips.shape
```

```python
# 두 columns에 대해 값을 각각 인덱스, 콜롬으로 하여 그 도수를 데이터프레임 형태로 반환
party_counts = pd.crosstab(tips["day"], tips["size"]) # crosstab..
party_counts
```

```python
# 재색인
party_counts = party_counts.reindex(index=["Thur", "Fri", "Sat", "Sun"], columns=[2,3,4,5])
party_counts
```

```python
# Normalize to sum to 1, 정규화
party_counts.sum(axis="columns")
```

```python
party_pcts = party_counts.div(party_counts.sum(axis="columns"), axis="index")
party_pcts # 요일마다 size= 2, 3, 4, 5값을 더함 (열방향), 각 행의 총합으로 나눔
```

```python
party_pcts.plot.bar(stacked=True)
```

  - hue 옵션과 더불어, 데이터프레임을 시각화하는데에 유용한 함수
  - catplot은 여러 subplots를 row와 columns 카테고리 기준에 따라 그림
  - kinds 옵션을 통해 그래프 종류를 결정

```python
fig, ax = plt.subplots(1, 2) # 하나의 figure 안에 두 개의 ax 만듦
sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5], ax=ax[0])
sns.barplot(x="day", y="tip_pct", hue="smoker", data=tips[tips.tip_pct < 1], ax=ax[1])
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 2, 1)
axes2 = fig.add_subplot(1, 2, 2)

sns.boxplot(x="tip_pct", y="day", data=tips[tips.tip_pct < 0.5], ax=axes1)
sns.barplot(x="day", y="tip_pct", hue="smoker", data=tips[tips.tip_pct < 1], ax=axes2)
```

### Catplot (categorical plot), row와 col로 category 결정

```python
# 어떤 그래프를 그릴지 kind 옵션으로 지정
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker", kind="bar", data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x="day", y="tip_pct", row="time", col="smoker", kind="bar", data=tips[tips.tip_pct < 1])
```
